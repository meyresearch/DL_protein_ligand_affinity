```python
import numpy as np
import ast,os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### KIBA Data

```python
path=os.getcwd()
```

```python
path1=path+'/Y_updated_kiba.npy'
y=np.load(path1)
affinity= y[~np.isnan(y)]
```

```python
sns.set(font_scale = 3,style="white")


fig = plt.figure(figsize=(15, 15))

s = pd.DataFrame({'val': affinity, 'Interaction':['Binding' if x<=3 else 'Non-binding' for x in affinity]})
ax=sns.histplot(data=s,bins=40, x="val", hue="Interaction",palette=['#EF9730','#047F8F'])
#ax =  sns.histplot(affinity,bins=20,color='#047F8F')
ax.set_xlabel('KIBA score',fontsize=35)
ax.set_ylabel('#Binding Interactions',fontsize=35)
ax.set_xlim([0,6])

# fig.savefig(os.path.join('results_data',  'Kiba_plot2' + '.png'), dpi=500, bbox_inches='tight')
# fig.savefig(os.path.join('results_data',  'Kiba_plot2' + '.pdf'), dpi=500, bbox_inches='tight')

```

### Davis Data

```python
path1=path+'/Y_updated_davis.npy'
y1=np.load(path1)
affinity1 = [-np.log10(y / 1e9) for y in y1.flatten()]
affinity1
```

```python
sns.set(font_scale = 3,style="white")
fig = plt.figure(figsize=(15, 15))

s = pd.DataFrame({'val': affinity1, 'Interaction':['Binding' if x>=7 else 'Non-binding' for x in affinity1]})
ax=sns.histplot(data=s,bins=5, x="val", hue="Interaction",palette=['#047F8F','#EF9730'])

ax.set_xlabel('$pK_d$',fontsize=35)
ax.set_ylabel('#Binding Interactions',fontsize=35)
#ax.set_xlim([,6])

# fig.savefig(os.path.join('results_data',  'Davis_plot2' + '.png'), dpi=500, bbox_inches='tight')
# fig.savefig(os.path.join('results_data',  'Davis_plot2' + '.pdf'), dpi=500, bbox_inches='tight')

```
