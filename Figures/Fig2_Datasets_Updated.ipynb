```python
import numpy as np
import ast,os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### KIBA Data

```python
path=os.getcwd()
```

```python
path1=path+'/Y_updated_kiba.npy'
y=np.load(path1)
affinity= y[~np.isnan(y)]
```

```python
sns.set(font_scale = 3,style="white")
```

```python
f, (ax_top, ax_bottom) = plt.subplots(ncols=1, nrows=2, sharex=True, gridspec_kw={'hspace':0.1},figsize=(15, 15))
s = pd.DataFrame({'val': affinity, 'Interaction':['Binding' if x<3 else 'Non-binding' for x in affinity]})
sns.histplot(data=s,bins=45, x="val", hue="Interaction",palette=['#EF9730','#047F8F'],ax=ax_top)
sns.histplot(data=s,bins=45, x="val", hue="Interaction",palette=['#EF9730','#047F8F'],ax=ax_bottom)

ax_top.set_ylim(bottom=15000,top=35000)   # those limits are fake
ax_bottom.set_ylim(0,7500)

sns.despine(ax=ax_bottom)
sns.despine(ax=ax_top, bottom=True)

ax_top.set_ylabel('', fontsize=25)
ax_bottom.yaxis.set_label_coords(-0.08, 0.5)
ax_bottom.set_yticks([0, 2500, 5000,7500])
ax_bottom.set_yticklabels(['0', '2500', '5000', '7500'], fontsize=35)

ax_bottom.set_xticks([0,1,2,3,4,5,6])
ax_bottom.set_xticklabels(['0','1','2','3','4','5','6'], fontsize=35)
ax_bottom.set_xlim([0,6])
ax_top.set_yticks([15000, 25000,35000])
ax_top.set_yticklabels(['15000', '25000','35000'], fontsize=35)
ax_bottom.set_xlim([0,6])
ax = ax_top
d = .045  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax_top.set_ylabel('')
# ax_top.
ax2 = ax_bottom
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal

ax2.set_xlabel('KIBA score',fontsize=35)
ax2.set_ylabel('',fontsize=25)
ax.set_ylabel('#Binding Interactions',fontsize=35,loc='bottom')
# #remove one of the legend
ax_bottom.legend_.remove()
ax_top.legend_.remove()
plt.show()

f.savefig(os.path.join('results_data',  'Kiba_plot_upd' + '.png'), dpi=500, bbox_inches='tight')
f.savefig(os.path.join('results_data',  'Kiba_plot_upd' + '.pdf'), dpi=500, bbox_inches='tight')
f.savefig(os.path.join('results_data',  'Kiba_plot_upd' + '.svg'), dpi=500, bbox_inches='tight')
```

### Davis Data

```python
path1=path+'/Y_updated_davis.npy'
y1=np.load(path1)
affinity1 = [-np.log10(y / 1e9) for y in y1.flatten()]
affinity1
```

```python
sns.set(font_scale = 2,style="white")
```

```python
f, (ax_top, ax_bottom) = plt.subplots(ncols=1, nrows=2, sharex=True, gridspec_kw={'hspace':0.1},figsize=(15, 15))
s = pd.DataFrame({'val': affinity1, 'Interaction':['Binding' if x>=7 else 'Non-binding' for x in affinity1]})
sns.histplot(data=s,bins=11, x="val", hue="Interaction",palette=['#047F8F','#EF9730'],ax=ax_top)
sns.histplot(data=s,bins=11, x="val", hue="Interaction",palette=['#047F8F','#EF9730'],ax=ax_bottom)

ax_top.set_ylim(bottom=10000,top=17000)   # those limits are fake
ax_bottom.set_ylim(0,1000)

sns.despine(ax=ax_bottom)
sns.despine(ax=ax_top, bottom=True)

ax_top.set_ylabel('', fontsize=25)
ax_bottom.yaxis.set_label_coords(-0.08, 0.5)
ax_bottom.set_yticks([0, 500, 1000])
ax_bottom.set_yticklabels(['0', '500', '1000'], fontsize=35)

ax_bottom.set_xticks([5,6,7,8,9,10,11])
ax_bottom.set_xticklabels(['5','6','7','8','9','10','11'], fontsize=35)
ax_top.set_yticks([5000, 10000,18000])
ax_top.set_yticklabels(['5000', '10000','18000'], fontsize=35)
# ax_bottom.set_xlim([5,10])
ax = ax_top
d = .045  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax_top.set_ylabel('')
# ax_top.
ax2 = ax_bottom
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal

ax2.set_xlabel('$pK_d$',fontsize=35)
ax2.set_ylabel('',fontsize=25)
ax.set_ylabel('#Binding Interactions',fontsize=35,loc='bottom')
# #remove one of the legend
ax_bottom.legend_.remove()
plt.show()

f.savefig(os.path.join('results_data',  'Davis_plot_upd' + '.png'), dpi=500, bbox_inches='tight')
f.savefig(os.path.join('results_data',  'Davis_plot_upd' + '.pdf'), dpi=500, bbox_inches='tight')
f.savefig(os.path.join('results_data',  'Davis_plot_upd' + '.svg'), dpi=500, bbox_inches='tight')
```
