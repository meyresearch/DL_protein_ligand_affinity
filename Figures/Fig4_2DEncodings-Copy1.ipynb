```python
from scipy.stats import wilcoxon
import pandas as pd

# Define the list of cmap methods to compare
cmap_list = ['ESM','Alphafold2',  'Pconsc4', 'Random']

# Create a dictionary to store the test results
test_results = {}

# Loop over each pair of cmap methods and perform Wilcoxon signed-rank test on each metric
for i in range(len(cmap_list)):
    for j in range(i+1, len(cmap_list)):
        cmap1 = cmap_list[i]
        cmap2 = cmap_list[j]
        for metric in ['CI', 'Pearson', 'Spearman', 'RMSE']:
            # Get the bootstrapped test performance data for cmap1 and cmap2
            data1 = df_melted1[(df_melted1['cmap']==cmap1) & (df_melted1['variable']==metric)]['value'].values
            data2 = df_melted1[(df_melted1['cmap']==cmap2) & (df_melted1['variable']==metric)]['value'].values
            # Perform Wilcoxon signed-rank test and store the p-value in the dictionary
            _, pvalue = wilcoxon(data1, data2)
            test_results[(cmap1, cmap2, metric)] = pvalue

# Convert the dictionary to a pandas DataFrame
test_results_df = pd.DataFrame(test_results.items(), columns=['Comparison', 'P-value'])

print("KIBA dataset results")
# Print the test results
print(test_results_df)
```

```python
from matplotlib.patches import PathPatch
import matplotlib.pyplot as plt
from matplotlib.gridspec import SubplotSpec

def create_subtitle(fig: plt.Figure, grid: SubplotSpec, title: str):
    "Sign sets of subplots with title"
    row = fig.add_subplot(grid)
    # the '\n' is important
    row.set_title(f'{title}\n', fontweight='bold',fontsize=35,ha='center',va='bottom')
    # hide subplot
    row.set_frame_on(False)
    row.axis('off')
    
def adjust_box_widths(g, fac):
    """
    Adjust the withs of a seaborn-generated boxplot.
    """

    # iterating through Axes instances
    for ax in g.axes:

        # iterating through axes artists:
        for c in ax.get_children():

            # searching for PathPatches
            if isinstance(c, PathPatch):
                # getting current width of box:
                p = c.get_path()
                verts = p.vertices
                verts_sub = verts[:-1]
                xmin = np.min(verts_sub[:, 0])
                xmax = np.max(verts_sub[:, 0])
                xmid = 0.5*(xmin+xmax)
                xhalf = 0.5*(xmax - xmin)

                # setting new width of box
                xmin_new = xmid-fac*xhalf
                xmax_new = xmid+fac*xhalf
                verts_sub[verts_sub[:, 0] == xmin, 0] = xmin_new
                verts_sub[verts_sub[:, 0] == xmax, 0] = xmax_new

                # setting new width of median line
                for l in ax.lines:
                    if np.all(l.get_xdata() == [xmin, xmax]):
                        l.set_xdata([xmin_new, xmax_new])
```
