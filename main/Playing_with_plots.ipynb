## Loading in data

```python
alpha =  pd.read_csv('plot1/alpha.csv')
esm = pd.read_csv('plot1/esm.csv')
random = pd.read_csv('plot1/random.csv')
pcons4 = pd.read_csv('plot1/pconsc4.csv')
```

```python
x = np.linspace(0,6,5)
```
